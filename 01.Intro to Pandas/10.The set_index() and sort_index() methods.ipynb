```python
import pandas as pd
```

```python
# reading and showing dataframe
df = pd.read_csv("StudentsPerformance.csv")
df.head(5)
```

# Create Index

```python
import numpy as np
import random
```

```python
# creating non repetitive values for the index
index = np.arange(0,1000)
```

```python
# shuffling the indexes
random.shuffle(index)
```

```python
# creating a new column with new index
df["random index"] = index
```

# Set Index

```python
# set new_index column as index
df.set_index("random index",inplace=True)
```

# Sort Index

```python
# sort dataframe by new index
df.sort_index()
```

```python
# sort descending by index
df.sort_index(ascending=False )
```
