```python
import pandas as pd
```

```python
# reading the csv file
df = pd.read_csv("StudentsPerformance.csv")
df.head(5)
```

# Math Operations

## Operations in columns

```python
# select a column and calculate total sum of the math score column
df["math score"].sum()
```

```python
# count, mean, std, max, and min
df["math score"].count() # count the number of the rows in the colomn
df["math score"].mean() #avrage val of the column
df["math score"].std() # standared divition
df["math score"].max() # max val 
df["math score"].min()  # min val 
```

```python
# easier calculation with .describe() can have all the methode of (count, mean, std, max, and min)
df["math score"].describe()
```

## Operations in rows

```python
# calculating the sum in a row of math score	reading score	writing score
df["sum"]=df["math score"] + df["reading score"] + df["writing score"]
df["sum"]
```

```python
# calculating the average score and assigning the result to a new column
df["avg"] = (df["math score"] + df["reading score"] + df["writing score"])/3
```
