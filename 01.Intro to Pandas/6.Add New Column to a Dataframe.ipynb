## Adding a new column with an array

The dataframe has 1000 rows, so we need to create an array with 1000 elements

```python
# import numpy
import numpy as np 
```

```python
# create an array of 1000 elements
verbal_score =np.arange(0,1000) 
```

```python
# length of the array
len(verbal_score)
```

```python
# adding a new column to dataframe with an array
df["verbal score"] = verbal_score
```

```python
# create random integer numbers between 50 and 100
language_score =  np.random.randint(50,100,size=1000)
```

```python
# min value inclusive and high value exclusive
min(language_score)
max(language_score)
```

```python
# adding a new column to dataframe with an array
df["language score"] = language_score
```
